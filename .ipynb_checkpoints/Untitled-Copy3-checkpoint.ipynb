```python
from model import BPR, FTD, HTD

from utils import train_dataset, test_dataset, evaluate, print_result, read_settings
from run import run_base, run_FTD, run_HTD
from copy import deepcopy
import time

import torch
import torch.utils.data as data
import torch.optim as optim

import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)

# Setup
## gpu setting
gpu = torch.device('cuda:2')

## Hyperparameters for Training
lr, batch_size = 0.001, 1024
reg = 0.001

## Load dataset
user_num, item_num = 5220, 25182

train_R, train_interactions, valid_R, test_R = read_settings()
train_dataset = train_dataset(user_num, item_num, train_R, train_interactions)
test_dataset = test_dataset(user_num, item_num, valid_R, test_R)
train_loader = data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

run_dict = {'train_loader':train_loader, 'test_dataset':test_dataset, 'lr':lr, 'reg':reg, 'gpu':gpu, \
            'early_stop':5, 'eval_period':4}
```

```python
## Generate Teacher model
model = BPR(user_num, item_num, dim=200, gpu=gpu).to(gpu)
Teacher_history = run_base(run_dict, model)
```

```python
## Generate Student model
model = BPR(user_num, item_num, dim=100, gpu=gpu).to(gpu)
Student_history = run_base(run_dict, model)
```

```python
## Load Trained Teacher
Teacher = BPR(user_num, item_num, dim=200, gpu=gpu)
with torch.no_grad():
    model_path = "./saved_models/200"
    Teacher = Teacher.to(gpu)
    Teacher.load_state_dict(torch.load(model_path))
    Teacher_user_emb, Teacher_item_emb = Teacher.get_embedding()
    del Teacher

run_dict['lmbda_TD'] = 0.001
model = FTD(user_num, item_num, Teacher_user_emb, Teacher_item_emb, gpu=gpu, student_dim=100).to(gpu)
FTD_history = run_FTD(run_dict, model)
```

```python
## Load Trained Teacher
Teacher = BPR(user_num, item_num, dim=200, gpu=gpu)
with torch.no_grad():
    model_path = "./saved_models/200"
    Teacher = Teacher.to(gpu)
    Teacher.load_state_dict(torch.load(model_path))
    Teacher_user_emb, Teacher_item_emb = Teacher.get_embedding()
    del Teacher

K = 30
design_choice = 'second'
run_dict['alpha'] = 0.5
model = HTD(user_num, item_num, Teacher_user_emb, Teacher_item_emb, gpu=gpu, student_dim=100, K=K, choice=design_choice).to(gpu)
HTD_history = run_HTD(run_dict, model)
```

```python
## Load Trained Teacher
Teacher = BPR(user_num, item_num, dim=200, gpu=gpu)
with torch.no_grad():
    model_path = "./saved_models/200"
    Teacher = Teacher.to(gpu)
    Teacher.load_state_dict(torch.load(model_path))
    Teacher_user_emb, Teacher_item_emb = Teacher.get_embedding()
    del Teacher

K = 30
design_choice = 'second'
run_dict['alpha'] = 0.
model = HTD(user_num, item_num, Teacher_user_emb, Teacher_item_emb, gpu=gpu, student_dim=100, K=K, choice=design_choice).to(gpu)
HTD_history2 = run_HTD(run_dict, model)
```
